```python
import cv2
import numpy as np
import os
import random
import csv
```

```python
video_names = ['videos/bookstore/video0/video.mov', 
               'videos/coupa/video1/video.mov', 
              'videos/deathCircle/video0/video.mov',
               'videos/gates/video4/video.mov',
               'videos/hyang/video6/video.mov',
               'videos/little/video0/video.mov',
               'videos/nexus/video3/video.mov',
               'videos/quad/video1/video.mov'
              ]
annotation_names = ['annotations/bookstore/video0/annotations.txt', 
                     'annotations/coupa/video1/annotations.txt',
                     'annotations/deathCircle/video0/annotations.txt',
                     'annotations/gates/video4/annotations.txt',
                     'annotations/hyang/video6/annotations.txt',
                     'annotations/little/video0/annotations.txt',
                     'annotations/nexus/video3/annotations.txt',
                     'annotations/quad/video1/annotations.txt'
                    ]
```

```python
train_imgs = 'imgs/train/'
val_imgs = 'imgs/val/'
train_ann = []
val_ann = []
```

```python
def do_work(image, annotation, frame_num, train, filename):
    global train_ann
    global val_ann
    
    valid_labels = ["Biker", "Pedestrian"]
    
    if train == 1:
        ## save in train folder
        cv2.imwrite(os.path.join(train_imgs, filename), image)
        path = train_imgs + str(filename)
    else:
        ## save in test folder
        cv2.imwrite(os.path.join(val_imgs, filename), image)
        path = val_imgs + str(filename)
    
    with open(annotation) as f:
        for line in f:
            vals = line.split()
            frame = vals[5]

            if int(frame) == frame_num:
                if int(vals[6]) == 0 and int(vals[7]) == 0: # not lost or occluded
                    entry = []
                    # add path to image
                    entry.append(path)
                    xmin = int(vals[1])
                    entry.append(xmin)
                    ymin = int(vals[2])
                    entry.append(ymin)
                    xmax = int(vals[3])
                    entry.append(xmax)
                    ymax = int(vals[4])
                    entry.append(ymax)
                    label = vals[-1]
                    label = label.replace('"','')
                    entry.append(label)
                    
                    print("Entry is: ", entry)
                    
                    if train == 1:
                        train_ann.append(entry)
                    else:
                        val_ann.append(entry)
```

```python
for i in range(len(video_names)):
    video_name = video_names[i]
    annotation = annotation_names[i]
    
    location = video_name.split('/')[1]
    video_num = video_name.split('/')[2]

    cap = cv2.VideoCapture(video_name)
    ret = True
    count = 0

    length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    print(length)

    while ret:
        ret, image = cap.read()
        count += 1
        
        if count % 50 == 0:
            seed = random.random()
            filename = str(location)+"_" +str(video_num)+"_"+str(count)+'.jpg'
            if seed <= 0.7 : 
                ## Train Set
                do_work(image, annotation, count, 1, filename)
            else:
                ## Val Set
                do_work(image, annotation, count, 0, filename)
```

```python
train_annotations_file = 'train_annotations.csv'

with open(train_annotations_file,'w+') as resultFile:
    wr = csv.writer(resultFile)
    for each in train_ann:
        wr.writerow(each)
```

```python
val_annotations_file = 'val_annotations.csv'

with open(val_annotations_file,'w+') as resultFile:
    wr = csv.writer(resultFile)
    for each in val_ann:
        wr.writerow(each)
```

```python
## write classes to csv
classes = []
for each in train_ann:
    classes.append(each[-1])
    
for each in val_ann:
    classes.append(each[-1])
```
